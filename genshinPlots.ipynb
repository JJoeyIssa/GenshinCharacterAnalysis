```python
#June 21, 2022
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
#opening the csv file
file = open('Genshin_Impact_All_Character_Stat.csv')
#making a file reader
csvReader = csv.reader(file)
#skip first line
next(csvReader)
#convert data into list
data = []
for row in csvReader:
    data.append(row)
```

```python
#create the empty list
dataChar = []
#you got to compare the list to something at first
prev = ""
#keep track of all the same indexes within the list
lastIndex = -1

for row in data:
    if row[0] != prev:
        #make a new list, wihtin the bigger list, since the name has changed
        dataChar.append([])
        #when time increases you know a new name appeared, at this index
        lastIndex += 1
        #this resets the cycle
        prev = row[0]
    #find the last index (where the new name appeared)
    #append that index (row) to the new list created in the for loop
    dataChar[lastIndex].append(row)
```

```python
#find and print certain information from each inner link

#for each row in the data set
for row in dataChar:
    #break it apart
    #for each entry in the index row
    for entry in row:
        #within the row, print the seperate entries
        print(entry)
    print("New Name")
        
```
